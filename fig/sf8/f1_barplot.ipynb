```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams.update({
    'pdf.fonttype': 42,
    'font.family': 'Arial',  # Set font to Arial
    'font.size': 14,  # Set font size to at least 14
    'axes.labelsize': 14,  # Axes labels size
    'axes.titlesize': 16,  # Title size for axes
    'xtick.labelsize': 14,  # X-tick label size
    'ytick.labelsize': 14,  # Y-tick label size
    'legend.fontsize': 14,  # Legend font size
    'figure.titlesize': 16,  # Figure title size
    'lines.linewidth': 2,  # Line width for clarity
    'axes.linewidth': 1.5,  # Axis line width
    'xtick.major.width': 1.5,
    'ytick.major.width': 1.5,
    'text.color': 'black',  # Make sure text is black (not grey)
    'axes.labelcolor': 'black',  # Ensure axes labels are black
    'axes.edgecolor': 'black',  # Ensure edge of axes is black
    'axes.facecolor': 'white',
    'xtick.color': 'black',  # Ensure X-tick marks are black
    'ytick.color': 'black',  # Ensure Y-tick marks are black
    'axes.titlepad': 20  # Padding for title
})
```

```python
x = np.arange(4) #["Virchow", "Virchow-2", "UNI", "UNI-2"]
y1 = [0.48, 0.55, 0.48, 0.49] #55
y2 = [0.55, 0.59, 0.56, 0.59] #100

width = 0.3

plt.figure(figsize= (5,3))
plt.bar(x-0.2, y1, width, color = 'grey')
plt.bar(x+0.2, y2, width, color = 'darkblue')
plt.xticks(x, ["Virchow", "Virchow-2", "UNI", "UNI-2"])
plt.xlabel("Foundation model")
plt.ylabel("F1 score")
plt.ylim(0,1)
plt.legend(["55 μm", "100 μm"], frameon = False,
           loc = 'upper right', bbox_to_anchor = (1.04,1.04))


# Add text labels above each bar
for i in range(len(x)):
    plt.text(x[i] - 0.2, y1[i] + 0.02, f"{y1[i]:.2f}", ha='center', va='bottom', fontsize=12)
    plt.text(x[i] + 0.2, y2[i] + 0.02, f"{y2[i]:.2f}", ha='center', va='bottom', fontsize=12)


plt.tight_layout()
plt.savefig("f1_score.pdf", bbox_inches = 'tight', transparent = True)
plt.show()
```
