```python
epoch_20_k_4 = np.load("epoch_20_k_4_results.npz")
epoch_30_k_4 = np.load("epoch_30_k_4_results.npz")
epoch_50_k_4 = np.load("epoch_50_k_4_results.npz")
```

```python
from sklearn.metrics import roc_curve, auc
fig, ax = plt.subplots(figsize=(4.9,4))
ax.set_aspect('equal', 'box') 

fpr, tpr, _ = roc_curve(epoch_20_k_4['all_labels'], epoch_20_k_4['all_outputs'])
plt.plot(fpr, tpr, color = 'b', alpha = 0.6, 
         label = f"20 ({auc(fpr, tpr):.3f})")

fpr, tpr, _ = roc_curve(epoch_30_k_4['all_labels'], epoch_30_k_4['all_outputs'])
plt.plot(fpr, tpr, color = 'r', alpha = 0.6,
         label = f"30 ({auc(fpr, tpr):.3f})")

fpr, tpr, _ = roc_curve(epoch_50_k_4['all_labels'], epoch_50_k_4['all_outputs'])
plt.plot(fpr, tpr, color = 'g', alpha = 0.6, 
         label = f"50 ({auc(fpr, tpr):.3f})")

plt.plot([0,1], [0,1], color = 'grey', linestyle = '--', lw = 1)

plt.legend(title = "Epochs (AUC)",     loc='center left',          # anchor legend by its left side
    bbox_to_anchor=(1, 0.5), frameon = False, handlelength = 0.4, handletextpad = 0.4)
plt.xlabel("FPR")
plt.xticks([0.00, 0.25, 0.50, 0.75, 1.00],
           ["0.00", "0.25", "0.50", "0.75", "1.00"],
           rotation=30, ha = "center")
plt.ylabel("TPR")
plt.yticks([0.00, 0.25, 0.50, 0.75, 1.00],
           ["0.00", "0.25", "0.50", "0.75", "1.00"],
           )
plt.title("k = 4", pad = 5)
plt.tight_layout()
plt.savefig("figs/roc_k_4.pdf", bbox_inches = 'tight', transparent = True)
plt.show()
```
