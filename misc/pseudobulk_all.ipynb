```python
h5ad_savedir = "h5ad"
os.makedirs(h5ad_savedir, exist_ok=True)
```

```python

pbs = []
for library_id in adata_vis.obs["library_id"].unique():
    lib_subset = adata_vis[adata_vis.obs['library_id']==library_id]
    rep_adata = sc.AnnData(X = np.asarray(lib_subset.X.sum(axis = 0)),
                           var = lib_subset.var[[]])
    rep_adata.obs_names = [library_id]
    for i in range(57):
        rep_adata.obs[lib_subset.obs.columns[i]] = lib_subset.obs[lib_subset.obs.columns[i]].iloc[0]
        
    # rep_adata.obs['cond'] = lib_subset.obs['cond'].iloc[0]
    # rep_adata.obs['seq_batch'] = lib_subset.obs['seq_batch'].iloc[0]
    # rep_adata.obs['outcome_24m'] = lib_subset.obs['outcome_24m'].iloc[0]

    pbs.append(rep_adata)

pb = sc.concat(pbs)
pb.write(os.path.join(h5ad_savedir,"pseudobulk_all.h5ad"))
```
